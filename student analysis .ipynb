```python

plt.hist(df['age'], bins=10, edgecolor='black')
plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.scatter(df['study_hours_per_day'], df['attendance_percentage'], alpha=0.6, color='green')
plt.title('Study Hours vs Attendance Percentage')
plt.xlabel('Study Hours per Day')
plt.ylabel('Attendance Percentage')
plt.grid(True)
plt.show()
```

```python
df['study_hours_group'] = pd.cut(df['study_hours_per_day'], bins=5)
grouped = df.groupby('study_hours_group')['attendance_percentage'].mean()

grouped.plot(kind='line', marker='o')
plt.title('Average Attendance vs Study Hour Groups')
plt.xlabel('Study Hour Groups')
plt.ylabel('Average Attendance')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['social_media_hours'], df['attendance_percentage'],alpha=0.6, color='red')
plt.title('Social Media Hours vs Attendance')
plt.xlabel('Social Media Hours per Day')
plt.ylabel('Attendance Percentage')
plt.show()
```

```python
df['diet_quality'].value_counts().plot(kind='bar', color='orange')
plt.title('Diet Quality Distribution')
plt.xlabel('Diet Quality')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.scatter(df['sleep_hours'], df['attendance_percentage'],alpha=0.6, color='orange')
plt.title('Sleep Hours vs Attendance')
plt.xlabel('Sleep Hours')
plt.ylabel('Attendance Percentage')
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(16, 4))

df.groupby('gender')['attendance_percentage'].mean().plot(kind='bar', ax=axs[0], color=['skyblue', 'lightgreen', 'orange'])
axs[0].set_title('Gender vs Attendance')

df.groupby('diet_quality')['sleep_hours'].mean().plot(kind='bar', ax=axs[1], color=['salmon', 'gold', 'lightgreen'])
axs[1].set_title('Diet vs Sleep Hours')

df.groupby('part_time_job')['attendance_percentage'].mean().plot(kind='bar', ax=axs[2], color=['red', 'green'])
axs[2].set_title('Part-Time Job vs Attendance')

plt.tight_layout()
plt.show()
```
